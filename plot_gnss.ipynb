```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pymap3d as pm
```

```python
df = pd.read_csv("/home/torsten/ms/data/rosbag2_2024_01_05-12_47_24/ublox.csv")
df
```

```python
%matplotlib tk
lat0 = df.iloc[0, 2]
lon0 = df.iloc[0, 3]
h0 = df.iloc[0, 4]

plt.figure()
plt.axis('equal')
plt.grid()
for index, row in df.iterrows():
    if index % 10 != 0:
        continue
    enu = pm.geodetic2enu(row["lat"], row["lon"], row["alt"], lat0, lon0, h0)
    plt.plot(enu[0], enu[1], "bx-")
    plt.errorbar(enu[0], enu[1], row["h_acc"], row["h_acc"])
    
```
